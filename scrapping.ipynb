```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
```

```python
url = requests.get('https://www.easy.com.ar/electrodomesticos/pequenos-electrodomesticos')
print(url)
```

```python
soup = BeautifulSoup(url.text, 'html.parser')
data = [
    json.loads(x.string) for x in soup.find_all("script", type="application/ld+json")
]
print(data[1]['itemListElement'])
```

```python
data_dt = []
item_list_element = data[1]['itemListElement']

for item in item_list_element:
    item_dict = item['item']
    offers_dict = item_dict.get('offers', {})

    data_dict = {
        'Nombre': item_dict.get('name'),
        'Precio': offers_dict.get('highPrice'),
        'Moneda': offers_dict.get('priceCurrency')
    }
    data_dt.append(data_dict)


df_a = pd.json_normalize(data_dt)
df_a
```
